```python
import pandas as pd
import re
from collections import Counter

# Load the dataset
df = pd.read_csv("cleaned_file.csv")

# Function to tokenize text by splitting on word boundaries
def custom_tokenize(text):
    return re.findall(r'\b\w+\b', str(text))  # Tokenizes based on word boundaries

# Tokenize the Tamil and Telugu columns
df['Tamil_tokens'] = df['Tamil'].apply(custom_tokenize)
df['Telugu_tokens'] = df['Telugu'].apply(custom_tokenize)

# Combine all tokens into lists for Tamil and Telugu
all_tamil_tokens = [token for tokens in df['Tamil_tokens'] for token in tokens]
all_telugu_tokens = [token for tokens in df['Telugu_tokens'] for token in tokens]

# Function to calculate bigram probabilities
def calculate_bigram_probabilities(tokens):
    # Count unigrams (single tokens)
    unigram_counts = Counter(tokens)
    
    # Count bigrams (pairs of consecutive tokens)
    bigram_counts = Counter(zip(tokens, tokens[1:]))
    
    # Calculate bigram probabilities
    bigram_probabilities = {}
    for (w1, w2), count in bigram_counts.items():
        bigram_probabilities[(w1, w2)] = count / unigram_counts[w1]
    
    return bigram_probabilities

# Function to create a table from bigram probabilities
def create_bigram_table(tokens, bigram_probabilities):
    unique_tokens = sorted(set(tokens))  # Unique words sorted alphabetically
    bigram_table = pd.DataFrame(0, index=unique_tokens, columns=unique_tokens)  # Initialize table with zeros
    
    # Fill the table with bigram probabilities
    for (w1, w2), prob in bigram_probabilities.items():
        bigram_table.loc[w1, w2] = prob
    
    return bigram_table

# Calculate bigram probabilities for Tamil and Telugu tokens
tamil_bigram_probabilities = calculate_bigram_probabilities(all_tamil_tokens)
telugu_bigram_probabilities = calculate_bigram_probabilities(all_telugu_tokens)

# Create tables for Tamil and Telugu bigram probabilities
tamil_bigram_table = create_bigram_table(all_tamil_tokens, tamil_bigram_probabilities)
telugu_bigram_table = create_bigram_table(all_telugu_tokens, telugu_bigram_probabilities)

# Save the tables to CSV files
tamil_bigram_table.to_csv("tamil_bigram_table.csv", encoding='utf-8-sig')
telugu_bigram_table.to_csv("telugu_bigram_table.csv", encoding='utf-8-sig')

print("\nBigram probability tables stored in tamil_bigram_table.csv and telugu_bigram_table.csv!")
```
